```python
# Collaborative Filtering
from sklearn.neighbors import NearestNeighbors
import pandas as pd

df = pd.read_csv('movie_ratings.csv', index_col='UserId')
df.fillna(0, inplace = True)
df
```

```python
neighbors = 3
knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
knn.fit(df.values)
distances, indices = knn.kneighbors(df.values, n_neighbors = neighbors)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files

# Upload file in Google Colab
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Function to process uploaded file and compute cosine similarity
def process_file(file_name):
    #df = pd.read_excel(file_name)
    df = pd.read_csv(file_name)
    df.rename(columns={df.columns[0]: 'Item'}, inplace=True)
    user_item_matrix = df.set_index("Item").T.fillna(0)
    cosine_sim_matrix = cosine_similarity(user_item_matrix.T)  # Compute item-item similarity
    item_sim_df = pd.DataFrame(cosine_sim_matrix, index=user_item_matrix.columns, columns=user_item_matrix.columns)
    return item_sim_df

# Compute similarity matrix
similarity_matrix = process_file(file_name)

# Extract similarity scores for movie
movie = 'm5'
if movie in similarity_matrix.index:
    movie_similarity = similarity_matrix.loc[movie].to_frame(name=f'Similarity with {movie}')
    print(f"Similarity of {movie} with other items:")
    display(movie_similarity)
else:
    print(f"{movie} not found in dataset.")
```
