```python
# prompt: drop 2 lakh datatset randomly

import pandas as pd
import random

# Assuming 'Astra' DataFrame is already loaded as in the provided code
# Astra = pd.read_csv('Astra.csv')

# Calculate the number of rows to drop
rows_to_drop = 300000

# Check if the number of rows to drop exceeds the total number of rows
if rows_to_drop >= len(Astra):
    print("Error: Cannot drop more rows than exist in the DataFrame.")
else:
    # Randomly select indices to drop
    indices_to_drop = random.sample(range(len(Astra)), rows_to_drop)

    # Drop the selected rows from the DataFrame
    Astra_dropped = Astra.drop(indices_to_drop)
Astra_dropped.shape
Astra.shape
    # Now Astra_dropped contains the dataset with 2 lakh rows dropped randomly
    # You can proceed with your further analysis or save this updated dataframe
    # Astra_dropped.to_csv("Astra_dropped.csv", index=False)
```

```python
G = nx.from_pandas_edgelist(Astra_dropped,
                            source='FromNodeId',
                            target='ToNodeId',
                            create_using=nx.DiGraph())
```

```python
print(G)
```

```python
nx.draw_networkx(G)
```

```python
degree_centrality = nx.degree_centrality(G)
degree_centralities = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)
for node, centrality in degree_centralities:
  print(f"Node: {node} Centrality: {centrality}")
```
