```python
G = nx.from_pandas_edgelist(iris,
                            source='sepal.length',
                            target='variety',
                            create_using=nx.DiGraph())
```

```python
print(G)
```

```python
nx.draw_networkx(G)
```

```python
degree_centrality = nx.degree_centrality(G)
degree_centralities = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)
for node, centrality in degree_centralities:
  print(f"Node: {node} Centrality: {centrality}")
```
